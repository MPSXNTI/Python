```python
#Estructura iterativas
#Ejercicios_10
#Pago bruto y liquido
while True:
    try:
        num_emp = int(input("Ingrese la cantidad de empleados: "))
        while num_emp <= 0:
            print("Error!")
            num_emp = int(input("Ingrese la cantidad de empleados: "))
        break
    except:
        print("Error!")

valor_hora = 1600
suma_b = 0
suma_l = 0
for i in range(1, num_emp + 1):
    while True:
        try:
            cant_horas = int(input("Ingrese las horas del empleado" + str(i) + ": "))
            while cant_horas <= 0:
                print("Error!")
                cant_horas = int(input("Ingrese las horas del empleado" + str(i) + ": "))
            break
        except:
            print("Error!")
    pago_b = valor_hora * cant_horas
    pago_l = int(pago_b * 0.8)
    print("El pago bruto del empleado", i, "es", pago_b)
    print("El pago líquido del empleado", i, "es", pago_l)
    suma_b = suma_b + pago_b
    suma_l = suma_l + pago_l

print("La empresa pago:", suma_b)
prom_emp = suma_l // num_emp
print("El pago promedio es:", prom_emp)
```

```python
#Estructura iterativas
#Ejercicios_11
#Calculadora de notas estudiantes
# Obtener el número de evaluaciones del usuario
num_evals = int(input("Profesor/a, ingrese el número de evaluaciones: "))

# Inicializar la suma total de las notas de los estudiantes y el número de estudiantes
total_grades = 0
num_students = 0

# Bucle para obtener las notas de los estudiantes y calcular el promedio
while True:
    # Obtener las notas del estudiante actual
    student_grades = []
    print("Notas para el estudiante", num_students + 1, ":")
    for i in range(num_evals):
        grade = float(input(f"Evaluación {i + 1} = "))
        student_grades.append(grade)

    # Calcular el promedio del estudiante
    student_average = sum(student_grades) / num_evals
    print("Notas para el estudiante:", student_average)

    # Actualizar la suma total de las notas y el número de estudiantes
    total_grades += student_average
    num_students += 1

    # Preguntar al usuario si desea ingresar las notas de otro estudiante
    continue_input = input("¿Desea ingresar las notas de otro estudiante?: ")
    if continue_input.upper() != "S":
        break

# Calcular e imprimir el promedio general del curso
course_average = total_grades / num_students
print("Promedio general del curso:", course_average)
```

```python
#Funciones, Listas y Archivos
#Ejercicios_16
#Encriptar palabra
def encriptar(palabra):
    """Encripta una palabra usando un algoritmo simple de desplazamiento.

    Args:
        palabra: La palabra a encriptar.

    Returns:
        La palabra encriptada.
    """
    palabra_encriptada = ''
    for letra in palabra:
        if letra.isalpha():
            # Desplaza la letra un carácter hacia adelante (A -> B, B -> C, etc.)
            codigo_ascii = ord(letra)
            if letra.isupper():
                codigo_ascii = (codigo_ascii - 65 + 1) % 26 + 65
            else:
                codigo_ascii = (codigo_ascii - 97 + 1) % 26 + 97
            palabra_encriptada += chr(codigo_ascii)
        else:
            # Si no es una letra, mantiene el carácter original
            palabra_encriptada += letra
    return palabra_encriptada

def contiene_espacios(palabra):
    """Verifica si una palabra contiene espacios en blanco.

    Args:
        palabra: La palabra a verificar.

    Returns:
        True si la palabra contiene espacios en blanco, False en caso contrario.
    """
    return ' ' in palabra

while True:
    palabra = input("Ingrese una palabra a encriptar: ")

    if contiene_espacios(palabra):
        print("La palabra contiene espacios en blanco. Inténtalo de nuevo.")
    elif len(palabra) > 30:
        print("La palabra es demasiado larga. Inténtalo de nuevo.")
    else:
        print("La palabra encriptada es:", encriptar(palabra))
        break
```
