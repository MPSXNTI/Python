Codificador de caracter normal a binario y viceversa

```python
"""
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

void codificarEnBinario(const char *cadena) {
    printf("Cadena en binario: ");
    for (int i = 0; i < strlen(cadena); i++) {
        char caracter = cadena[i];
        for (int j = 7; j >= 0; j--) {
            printf("%d", (caracter >> j) & 1);
        }
        printf(" ");
    }
    printf("\n");
}

void descodificarEnBinario(const char *cadenaBinaria) {
    printf("Cadena descodificada: ");
    char byte = 0;
    for (int i = 0; i < strlen(cadenaBinaria); i++) {
        if (cadenaBinaria[i] == ' ' || cadenaBinaria[i] == '\n') {
            printf("%c", byte);
            byte = 0;
        } else {
            byte = (byte << 1) | (cadenaBinaria[i] - '0');
        }
    }
    printf("\n");
}

int main() {
    int opcion;
    char cadena[100];

    do {
        printf("\nMenú:\n");
        printf("1. Codificar en binario\n");
        printf("2. Descodificar en binario\n");
        printf("3. Salir\n");
        printf("Seleccione una opción: ");
        scanf("%d", &opcion);

        switch (opcion) {
            case 1:
                printf("Ingrese la cadena de caracteres a codificar en binario: ");
                scanf("%s", cadena);
                codificarEnBinario(cadena);
                break;
            case 2:
                printf("Ingrese la cadena de binario a descodificar: ");
                getchar();  // Consumir el carácter de nueva línea pendiente
                fgets(cadena, sizeof(cadena), stdin);
                descodificarEnBinario(cadena);
                break;
            case 3:
                printf("Saliendo del programa. ¡Hasta luego!\n");
                break;
            default:
                printf("Opción no válida. Por favor, seleccione una opción válida.\n");
        }

    } while (opcion != 3);

    return 0;
}
"""
```
